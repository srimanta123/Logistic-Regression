```python
from sklearn.impute import SimpleImputer #most frequent data here is the mode
imputer = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
x = imputer.fit_transform(x)
y = imputer.fit_transform(y)
```

```python
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Label1=LabelEncoder()
x[:,5]=Label1.fit_transform(x[:,5])

Label2=LabelEncoder()
x[:,7]=Label2.fit_transform(x[:,7])

Label3=LabelEncoder()
x[:,8]=Label3.fit_transform(x[:,8])

Label4=LabelEncoder()
x[:,-2]=Label4.fit_transform(x[:,-2])

Label6=LabelEncoder()
y=Label6.fit_transform(y)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(x_train)
```

```python
print(y_train)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
#y_predict=model.predict(x_test)
model_score=model.score(x_test,y_test)
print(model_score)
```
